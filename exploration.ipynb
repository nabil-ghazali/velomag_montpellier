```python
import requests
import json

POST_API_URL = "https://api.example.com/submit_prediction" 

def post_prediction(compteur_id, prediction_value):
    
    # 1. Définir les données à envoyer (généralement au format JSON)
    payload = {
        "compteur_ref": compteur_id,
        "date_prevision": "2025-11-26T00:00:00Z",
        "valeur_predite": prediction_value,
        "unite": "velos/heure"
    }

    # 2. Définir les headers pour indiquer que le contenu est du JSON
    headers = {
        "Content-Type": "application/json"
    }
    
    try:
        # 3. Envoyer la requête POST
        response = requests.post(
            POST_API_URL, 
            data=json.dumps(payload), # Convertit le dictionnaire Python en chaîne JSON
            headers=headers
        )

        response.raise_for_status()
        
        print("\nSoumission de prédiction réussie !")
        print(f"Statut de la réponse : {response.status_code}")
        # Si l'API retourne un message de confirmation
        # print(response.json()) 

    except requests.exceptions.RequestException as e:
        print(f"Échec de la soumission POST : {e}")

if __name__ == "__main__":
    # Simuler l'envoi d'une prédiction
    post_prediction("TC01", 1500)
```
